```python
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),
                                     inplace=True)

## DataFrame.method(inplace=True) → usually fine.
## DataFrame["column"].method(inplace=True) → can trigger chained assignment warnings.
```

```python
car_sales_missing["Odometer"] = car_sales_missing["Odometer"].fillna(
    car_sales_missing["Odometer"].mean()
)
```

```python
# Column for series
seats_column = pd.Series([5,5,5,5,5])

# New column called seats 
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"] = car_sales["Seats"].fillna(5)
car_sales
```

```python
# Column from Python List 
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["fuel per 100km"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"]/100 * car_sales["fuel per 100km"]
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed road saftey"] = True
car_sales.dtypes
```

```python
# Drop a particular column
car_sales = car_sales.drop("Total fuel used (L)", axis = 1)
car_sales
```
